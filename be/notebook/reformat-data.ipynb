```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/Training.csv")
```

```python
df.head(5)
```

```python
symptoms = df.drop(columns="prognosis").columns
```

```python
new_feature = []
```

```python
max = 0
for row in df.values[:,:-1]:
    if sum(row) > max:
        max = sum(row)
print(max)
```

```python
for i in range(1, max + 1):
    new_feature.append("feature_{}".format(i))
```

```python
new_feature.append("disease")
```

```python
new_train = pd.DataFrame(columns=new_feature)
```

```python
new_train.head()
```

```python
row1 = df.values[0][:-1]
row1
```

```python
np.where(row1)
```

```python
symptoms[np.where(row1)]
```

```python
count = 0
for row in df.values:
    s = list(symptoms[np.where(row[:-1])])
    d = row[-1]
    val = s + [None] * (max - len(s)) + [d]
    new_train.loc[count] = val
    count += 1
```

```python
new_train.head(5)
```

```python
new_train.to_csv("../data/new_train.csv", index=False)
```

```python
test = pd.read_csv("../data/Testing.csv")
symp = test.columns
count = 0
new_test = pd.DataFrame(columns=new_feature)
for row in test.values:
    s = list(symp[np.where(row[:-1])])
    d = row[-1]
    val = s + [None] * (max - len(s)) + [d]
    new_test.loc[count] = val
    count += 1
```

```python
new_test.head(5)
```

```python
new_test.to_csv("../data/new_test.csv", index=False)
```
